```python
import matplotlib.pyplot as plt
import numpy as np
import os

data = np.loadtxt('../Outputs/NPT-1.ener',skiprows=1)

U = data[:,4]

time = data[:,1] #in fs

plt.figure()
plt.plot(time, U, color='green')
plt.xlabel("Time, fs")
plt.ylabel("Energy, a.u.")
plt.title('NVT Potential energy')
plt.show()

plt.savefig('../Figures/NVT_energy.png')

data = np.loadtxt('../Outputs/NVT-1.ener',skiprows=1)

U = data[:,4]

time = data[:,1] #in fs

plt.figure()
plt.plot(time, U, color='red')
plt.xlabel("Time, fs")
plt.ylabel("Energy, a.u.")
plt.title('NPT Potential energy')
plt.show()

plt.savefig('../Figures/NPT_energy.png')
```
