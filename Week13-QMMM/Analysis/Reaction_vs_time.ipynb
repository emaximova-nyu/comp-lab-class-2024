```python
import matplotlib.pyplot as plt
import numpy as np
import os

data = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog',skiprows=1)

q = data[:,1]

time = data[:,0] #in fs

# The reaction coordinate monitored in this run is the distance between d(C5663 and H5675) and d(C5670 and H5671) 

plt.figure()
plt.plot(time, q, color='green', label = 'Standard QM/MM')


data = np.loadtxt('../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog',skiprows=1)

q = data[:,1]

time = data[:,0] #in fs

plt.plot(time, q, color='Red', label = 'Metadynamics 1, height=0.5, width=0.5, freq=1/100')

# The second metadynamics run has hills spawned more frequently (every 20 steps instead of 100). 
# The sampling is faster but the ligand is ripped apart at the end.

data = np.loadtxt('../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog',skiprows=1)

q = data[:,1]

time = data[:,0] #in fs

plt.plot(time, q, color='Purple', label = 'Metadynamics 2, height=0.5, width=0.5, freq=1/20')
plt.xlabel("Time, fs")
plt.ylabel("Distance, A")
plt.title('Reaction coordinate vs time')
plt.legend()
plt.show()

plt.savefig('../Figures/reaction_coord.png')
```
