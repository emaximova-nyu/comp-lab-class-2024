```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = md.load_xtc('trp-cage/md_fit_center_no_solv_concat.xtc',top='trp-cage/zero_frame.pdb')
```

```python
end_to_end_dist = md.compute_distances(t, [[0,500]], periodic=True)
#print(end_to_end_dist)

rg = md.compute_rg(t)
#print(rg)
```

```python
plt.figure()
plt.plot(np.array(t.time)/1000,np.array(rg),'-b',label='Radius of Gyration [nm]')
plt.xlabel('Time [ns]')
plt.ylabel('Radius of Gyration [nm] ')
plt.ylim([0.9,1.15])
plt.savefig('../Figures/trp-cage_Rg.jpg')

plt.figure()
plt.plot(np.array(t.time)/1000,np.array(end_to_end_dist),'-r',label='End-to-end Distance [nm]')
plt.xlabel('Time [ns]')
plt.ylabel('End-to-end Distance [nm]')
plt.ylim([0,2.5])
plt.savefig('../Figures/trp-cage_end-to-end_dist.jpg')
```
