```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np

t_1hz3 = md.load_xtc('1hz3_T310.stepid25000000.every100ps.nowater.xtc',top='1hz3_T310.start.nowater.gro')

end_to_end_dist = md.compute_distances(t_1hz3, [[0,407]], periodic=True)
print(end_to_end_dist)

rg = md.compute_rg(t_1hz3)
print(rg)
```

```python
plt.figure()
plt.plot(np.array(t_1hz3.time)/1000,np.array(rg),'--bo',label='Radius of Gyration [nm]')
plt.plot(np.array(t_1hz3.time)/1000,np.array(end_to_end_dist),'--ro',label='End-to-end Distance [nm]')
plt.xlabel('Time [ns]')
plt.ylabel('End-to-end Distance, Radius of Gyration [nm] ')
plt.ylim([0,5])
plt.legend()
```

```python
plt.figure()
plt.hist(np.array(end_to_end_dist),bins=10,color='red',density=True)
plt.xlabel('End-to-end Distance [nm]')
plt.ylabel('Density')

plt.figure()
plt.hist(np.array(rg),bins=15,color='blue',density=True)
plt.xlabel('Radius of Gyration [nm]')
plt.ylabel('Density')
```
