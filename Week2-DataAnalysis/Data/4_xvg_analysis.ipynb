```python
file = open("1hz3_T310.run.25000000.energy.xvg","r")
data = file.readlines()[29:]

data_t = []
data_1 = []
data_2 = []
data_3 = []
data_4 = []
data_5 = []
data_6 = []

data_all = []

for i in data:
    data_i_split_raw = i.split('  ')
    data_i_split = []

    for k in data_i_split_raw:
        if k != '':
            data_i_split.append(float(k))
    #print(data_i_split)

    data_t.append(data_i_split[0])
    data_1.append(data_i_split[1])
    data_2.append(data_i_split[2])
    data_3.append(data_i_split[3])
    data_4.append(data_i_split[4])
    data_5.append(data_i_split[5])
    data_6.append(data_i_split[6])
 
data_all.append(data_1)
data_all.append(data_2)
data_all.append(data_3)
data_all.append(data_4)
data_all.append(data_5)
data_all.append(data_6)

#print(data_all[5])
```

```python
import matplotlib.pyplot as plt
import numpy as np

ylabels = ["Potential [kJ/mol]", "Kinetic En. [kJ/mol]", "Total Energy [kJ/mol]", "Temperature [K]", "Pressure [bar]", "Volume [nm^3]"]

for i in range(6):
    plt.figure()
    plt.plot(np.array(data_t)/1000,np.array(data_all[i]),color = np.random.rand(3,))
    plt.xlabel('Time [ns]')
    plt.ylabel(ylabels[i])
```

```python
fig, ax1 = plt.subplots() 
 
ax1.set_xlabel('Time [ns]') 
ax1.set_ylabel('Temperature [K]',color = 'red') 
ax1.plot(np.array(data_t)/1000, np.array(data_all[2]), color = 'red',label='T') 
ax1.tick_params(axis ='y') 
 
ax2 = ax1.twinx() 
 
ax2.set_ylabel('Kinetic Energy [kJ/mol]',  color = 'green', rotation=270, labelpad=15) 
ax2.plot(np.array(data_t)/1000, np.array(data_all[4]), color = 'green', label=r'E_{K}') 
ax2.tick_params(axis ='y') 
```
